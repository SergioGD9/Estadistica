```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Creamos la variable

```python
ventas = [50,60,70,80,90]
gastos_publicidad = [30,40,50,60,70]
```

### Creamos el coeficiente de correlación de Pearson

```python
coef_correlacion = np.corrcoef(ventas, gastos_publicidad)[0,1]
print(f'Coeficiente de correlación de Pearson {coef_correlacion}')
```

### Modificamos las variables para hacer la regresión lineal

```python
ventas = np.array([50,60,70,80,90])
gastos_publicidad = np.array([30,40,50,60,70])
```

### Creamos el modelo de regresión lineal

```python
model = LinearRegression()
```

### Ajustamos el modelo

```python
model.fit(gastos_publicidad.reshape(-1,1),ventas)
```

### Obtenemos el coeficiente de la pendiente y la intersección

```python
pendiente = model.coef_[0]
interseccion = model.intercept_
```

### Generamos las predicciónes

```python
ventas_pred = model.predict(gastos_publicidad.reshape(-1,1))
print(ventas_pred)
```

### Graficamos

```python
plt.scatter(gastos_publicidad,ventas,label='Ventas sobre gastos publicidad')
plt.plot(gastos_publicidad,ventas_pred,color='red',label='Regresión Lineal')
plt.xlabel('Gastos Publicidad')
plt.ylabel('Ventas')
plt.title('Ventas sobre gastos publicidad')
plt.legend()
plt.show()
```

### Mostramos valores

```python
print(f'Pendiente: {pendiente}')
print(f'Intersección: {interseccion}')
```
