# 01 Histograma

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
edades = [22,31,25,35,21,19,22,32,19,32,31,29,21,29,19,24,26,25,18,30,21]
```

```python
#Histograma
plt.hist(edades, bins=10, edgecolor='black')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')#Frecuencia con la que aparece una edad
plt.title('Histograma de edades de estudiantes')
plt.show()
```

```python
# Guardamos la figura en un archivo
plt.savefig('histograma.png')
```
