```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

### Vamos a crear el espacio muestral en el que calcularemos la probabilidad en el que dos dados con seis caras hagan una suma igual o mayor a 8

1. Espacio muestral

```python
espacio_muestral = [(i,j) for i in range(1,7) for j in range(1,7)]
print(espacio_muestral)#Mostramos todas las combinaciones posibles
```

2. Obtenemos el evento A que representa una suma mayor o igual a 8

```python
evento_A = [(i,j) for i, j in espacio_muestral if i + j >=8]
print(evento_A)#Mostramos las combinaciones que dan 8 o más de 8
```

3. Calculamos la probabilidad

```python
probabilidad_A = (len(evento_A)/len(espacio_muestral))*100
probabilidad_A_redondeada = round(probabilidad_A, 2)
print(f'La probabilidad de obtener valores mayores o iguales a 8 es de: {probabilidad_A_redondeada}%')
```

4. Obtener número par con el primer dado

```python
evento_b = [(i,j) for i, j in espacio_muestral if i%2 == 0]
print(evento_b)
```

5. Probabilidad de evento B

```python
probabilidad_B = len(evento_b)/len(espacio_muestral)*100
probabilidad_B_redondeada = round(probabilidad_B,2)
print(f'La probabilidad de obtener números pares con el primer dado es de: {probabilidad_B_redondeada}')
```

6. Graficos de dispersión

```python
x_A, y_A = zip(*evento_A)
x_B, y_B = zip(*evento_b)
```

```python
plt.scatter(x_A,y_A, color='blue', label='Evento A: suma >=8')
plt.scatter(x_B,y_B, color = 'red', label= 'Evento B: Número par en el primer dado')
plt.xlabel('Resultado Dado 1')
plt.ylabel('Resultadi Dado 2')
plt.title('Eventos A y B en el lanzamiento de Dados')
plt.legend()
plt.show()
```
